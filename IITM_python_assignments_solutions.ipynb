```python
def merge(D1, D2, priority):
  if priority == 'first':
    D_P, D_nP = D1, D2
  if priority == 'second':
    D_P, D_nP = D2, D1
  D = {}
  for i in D_nP.keys():
    D[i] = D_nP[i]
  for i in D_P.keys():
    D[i] = D_P[i]
  return(D)
```
